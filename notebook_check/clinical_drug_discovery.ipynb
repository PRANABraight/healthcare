```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set a professional plot style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries imported successfully.")
```

```python
# Load the dataset
try:
    df = pd.read_csv('C:/Users/prana/OneDrive/Desktop/4trimester/healthcare/data/Clinical Data_Discovery_Cohort.csv')
    print("Dataset loaded successfully.")
    print(f"Shape of the data: {df.shape}")
except FileNotFoundError:
    print("Error: 'Clinical Data_Discovery_Cohort.csv' not found. Please ensure it's in the same directory.")
```

```python
# Display the first 5 rows
print("--- First 5 Rows ---")
print(df.head())

# Get a summary of the dataframe, including data types and non-null counts
print("\n--- Dataframe Info ---")
df.info()

# Check for the number of missing values in each column
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# Clean and standardize column names
original_columns = df.columns
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('%', 'percent')
new_columns = df.columns

print("--- Column Names Cleaned ---")
print("Original:", list(original_columns))
print("Cleaned: ", list(new_columns))
```

```python
# Get descriptive statistics for numerical columns
print("--- Descriptive Statistics ---")
print(df.describe().round(2))
```

```python
# Distribution of Patient Age
plt.figure(figsize=(10, 6))
sns.histplot(df['dead_or_alive'], bins=20, kde=True)
plt.title('Distribution of Patient Alive or Dead', fontsize=16)

plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Gender
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', data=df, palette='viridis')
plt.title('Gender Distribution in the Cohort', fontsize=16)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
